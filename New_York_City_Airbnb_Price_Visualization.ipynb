## New York City Airbnb Price Visualization
Dataset used: https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing = pd.read_csv("AB_NYC_2019.csv")
housing.head(3)
```

```python
housing.info()
```

```python
housing['neighbourhood_group'].value_counts()
```

```python
housing.describe()
```

```python
housing.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
housing['number_of_reviews'].hist(bins=50, range=(0,100), figsize=(5, 3))
```

```python
housing['price'].hist(bins=50, range=(0,100), figsize=(5, 3))
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
            s=housing['price']/100, label="price", figsize=(12,8),
            c="price", cmap=plt.get_cmap("jet"), sharex=False)

plt.legend()
```

```python
bimport matplotlib.image as mpimg

housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
            s=housing['price']/100, label="price", figsize=(12,8),
            c="price", cmap=plt.get_cmap("jet"), sharex=False)

nyc_img = mpimg.imread("nyc.png")

plt.imshow(nyc_img, extent=[-74.25, -73.69, 40.49, 40.92], alpha=0.5, cmap=plt.get_cmap("jet"))

plt.xlabel("Longitude", fontsize=14)
plt.ylabel("Latitude", fontsize=14)

plt.legend(fontsize=14)
plt.show()
```
